## Install SSH Server

Before setting up the SSH, 

Step 1: start by going to your terminal and generate an SSH key -> 
Windows: `ssh-keygen -t rsa -b 4096 -C "kaggle_remote_ssh" -f .ssh/kaggle_rsa`
WSL: `ssh-keygen -t rsa -b 4096 -C "kaggle_remote_ssh" -f ~/.ssh/kaggle_rsa`

Step 2: Run `cat ~/.ssh/kaggle_rsa.pub` and copy the key

Step 3: Run `echo 'REPLACE_IT_WITH_YOUR_COPIED_KEY_FROM_STEP_2' >> ~/.ssh/authorized_keys`

Step 3: Create a Github repo and put the file named `authorized_keys` in it.

Step 4: go to the file and click raw and copy the url

```python
# Step 5: Copy the url and run -> https://raw.githubusercontent.com/0xpix/SSH_Key_public/refs/heads/main/authorized_keys
!bash install_ssh_server.sh "authorized_keys_URL_FROM_STEP_4"
```
